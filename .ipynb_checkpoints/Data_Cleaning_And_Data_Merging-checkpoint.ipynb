```python
annotation_df = pd.read_csv(r"C:\Users\gargi\Desktop\Publication\Transcriptomic_and_Epigenomic_Features\bwMtrPidtktV3hgO.csv")  

annotation_df.head()
```

```python
# Summary
print(annotation_df.info())
```

```python
# Standardize rsID column names
gwas_df.rename(columns={"rsID": "Uploaded_variation"}, inplace=True)
diagram_df.rename(columns={"SNP": "Uploaded_variation"}, inplace=True)
```

```python
diagram_df.rename(columns={"SNP": "Uploaded_variation"}, inplace=True)
```

```python
# Merge GWAS and DIAGRAM
merged_1 = pd.merge(gwas_df, diagram_df, on="Uploaded_variation", how="inner")
```

```python
# Merge with functional annotations
merged_final = pd.merge(merged_1, annotation_df, on="Uploaded_variation", how="inner")
```

```python
# Drop duplicates
merged_final.drop_duplicates(inplace=True)
```

```python
# Filter for genome-wide significance
if 'P' in merged_final.columns:
    merged_final = merged_final[merged_final['P'] < 5e-8]
```

```python
# Save cleaned and merged data
merged_final.to_csv("cleaned_merged_gwas_functional.csv", index=False)

# Preview result
print(merged_final.shape)
print(merged_final[['Uploaded_variation', 'SYMBOL', 'Consequence', 'P']].head())
```
