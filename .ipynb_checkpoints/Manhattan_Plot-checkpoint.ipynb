```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Use your merged_final dataset
df = merged_df.copy()

# Use DIAGRAM coordinates
df['CHROMOSOME'] = pd.to_numeric(df['CHROMOSOME_y'], errors='coerce')
df['POSITION'] = pd.to_numeric(df['POSITION_y'], errors='coerce')
df = df.dropna(subset=['CHROMOSOME', 'POSITION', 'P_VALUE'])

# Calculate -log10(P)
df['-log10(P)'] = -np.log10(df['P_VALUE'])

# Sort by chromosome and position
df = df.sort_values(['CHROMOSOME', 'POSITION'])

# Create index and color
df['ind'] = range(len(df))
df['color'] = ['#1f77b4' if i % 2 == 0 else '#ff7f0e' for i in df['CHROMOSOME']]

# Plot
plt.figure(figsize=(14, 6))
plt.scatter(df['ind'], df['-log10(P)'], c=df['color'], s=20, edgecolor='none')
plt.axhline(y=-np.log10(5e-8), color='red', linestyle='--', linewidth=1)

# Chromosome ticks
ticks = df.groupby('CHROMOSOME')['ind'].median()
plt.xticks(ticks, ticks.index)
plt.xlabel('Chromosome')
plt.ylabel('-log10(P)')
plt.title('Manhattan Plot of GWAS Significance')
plt.tight_layout()
plt.show()

```
